# Datathon

## Desafio

Marta, gerente do centro cirúrgico, fez um levantamento de todas as cirurgias realizadas no Hospital Santa Cruz no ano de 2018 e início de 2019. Dados como perfil de pacientes, cirurgias mais realizadas e tempo médio das cirurgias por especialidade são muito importantes para a área e no momento eles não dispõem de nenhuma interface para consulta destes indicadores. A área disponibilizou o conjunto de dados em um arquivo colunar e solicitou que sua equipe a auxilie com uma análise geral desta informação. 

Ela também está interessada em automatizar a classificação do porte das cirurgias e solicitou para seu grupo a POC de um modelo capaz de classificar o porte levando em consideração os dados fornecidos.

Para auxiliar os times de desenvolvimento, a área também forneceu um catálogo descritivo dos campos existentes no arquivo:

| Coluna                       |                                                                               Descrição |
|------------------------------|-----------------------------------------------------------------------------------------|
| Fonte pagadora               |                          Plano de saúde responsável pelo pagamento da conta do paciente |
| Tipo de passagem             |                     Refere-se ao tipo de admissão do paciente (internado ou emergência) |
| Tipo de procedimento         |                                                            Agendado/Eletivo ou Urgência |
| Porte                        | Está relacionado a complexidade da cirurgia. Quanto maior o porte, maior a complexidade |
| Tempo de permanencia         |                                         Tempo em que o paciente ficou em sala cirúrgica |
| Data e hora registrorpapreop |                               Data e hora em que foi iniciado o registro pré-operatório |
| RPA                          |                                                              Recuperação pós-anestésica |
| ATB                          |                                                                             Antibiótico |
| ASA cirurgico                |                                                                Indica o risco cirúrgico |
| PAS                          |                                                              Pressão arterial sistólica |
| PAD                          |                                                             Pressão arterial diastólica |
| SPO                          |                                                                   Saturação do paciente |
| FC                           |                                                                     Frequência cardíaca |
| RA                           |                                                                  Recuperação anestésica |
| CID                          |                                                         Código internacional de doenças |
| Deltatempo                   |                                                                       Tempo da cirurgia |

Lembre que, não necessariamente, o escopo deste projeto deve ser fechado nas sugestões apontadas pela Marta. Sua equipe tem liberdade para explorar estes dados para desenvolvimento de indicadores de valor para tomadas de decisão na área.

Boa sorte!

Os alunos tem a liberdade de realizar a atividade de forma individual ou formar grupos de até 5 pessoas. O prazo para entrega do material totaliza 15 dias corridos considerando o início do prazo no término da aula sobre o Datathon.

As atividades devem ser entregues por e-mail contendo no mínimo dois arquivos:

* Uma apresentação do trabalho (ppt) mostrando as etapas de processamento dos dados e os raciocínios empregados na criação dos indicadores. Caso a interface escolhida seja o PowerBI, incluir imagem das telas criadas através da plataforma.
* O Notebook analítico com código e resultados obtidos no treinamento do modelo.
