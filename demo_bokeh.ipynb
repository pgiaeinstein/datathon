```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show

import numpy as np

output_notebook()

valores_x = [1, 2, 3, 4, 5]

y   = [x+2 for x in valores_x] # x+2
y_2 = np.log(x) # log(x)
y_3 = [x**2 for x in valores_x] # x^2
y_4 = [(-1)**x for i, x in enumerate(valores_x)] # -1^x

p = figure(title="Linha", x_axis_label='eixo x', y_axis_label='eixo y')

p.line(x, y, legend="Função: x+2", line_width=2, line_color='red', line_dash=[2])
p.line(x, y_2, legend="Função: log(x)", line_width=2, line_color='green')
p.line(x, y_3, legend="Função: x^2", line_width=2, line_color='black', line_join='miter')
p.line(x, y_4, legend="Função: -1^x", line_width=1, line_color='blue', line_join='miter')

show(p)
```

```python
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [5, 3, 4, 2, 4, 6]

p = figure(x_range=fruits, plot_height=250, title="Fruit Counts", toolbar_location=None)

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```
