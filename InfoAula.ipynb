# Datathon

## Desafio

Marta, gerente do centro cirúrgico, fez um levantamento de todas as cirurgias realizadas no Hospital Santa Cruz no ano de 2018. Ela disponibilizou estes dados em um arquivo csv e solicitou que sua equipe a auxilie com uma análise geral desta informação. Dados como perfil de pacientes, cirurgias mais realizadas, tempo médio das cirurgias por especialidade são muito importantes para a área.

Ela também está interessada em automatizar a classificação do porte das cirurgias e solicitou para seu grupo uma POC de um modelo capaz de classificar o porte levando em consideração os dados disponibilizados.

Boa sorte!
